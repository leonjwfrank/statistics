```python
# 生成器
def gen_fb(n):
	a,b = 0,1
	if n in [1, 2]:
		yield 1
	while True:
		x = a + b
		a = b
		b = x
		yield x
```

```python
fbn = gen_fb(1)
print(type(fbn))
print(type(list(fbn)))
```

```python
for i in range(10):
    print(next(fbn))
```

```python
import functools
import profile

#  递归， 默认最多1000层
@functools.lru_cache(maxsize=None)
def fb(n):
	if n <= 2:
		return 1
	else:
		return fb(n-1) + fb(n-2)
```

```python
if '__name__' == '__main__':
    for i in range(10):
        print('ind:',i)
        profile.run("print('value', fb(i));print()")
```
